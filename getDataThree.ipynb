```python
col_name = df1.keys()[3]               #name
amount_col_name = df1.keys()[8]        #All aggregations and total
piece_count_col_name = df1.keys()[6]   #no. of pcs.
```

```python
#Series containing the Aggregations

myseries = df1[col_name]
myseries

def find(s, el):
    for i in s.index:
        if s[i] == el: 
            return i
    return None

index = find(myseries, "TOTAL NO PCS.")
print("Finding the Index of Aggregations ....")
if(index != None):
    print(index)
else:
    print("Index Not found, check the column number !")
```

```python
def build_dict(bill, df, index):
    Dictonary = {
    bill :{    
        "TOTAL NO PCS.": df[piece_count_col_name][index],		
        "TOTAL AMOUNT": df[amount_col_name][index + 1],		
        "(मूर्ति)  दुकान":	df[amount_col_name][index + 2],	
        "(गणेश लक्ष्मी)  दुकान": df[amount_col_name][index + 3],		
        "LOADING CHARGE (लेबर खर्चा)": df[amount_col_name][index + 4],		
        "TRANSPORTATION (गाड़ी भाड़ा)": df[amount_col_name][index + 5],		
        "PACKING CHARGES (कार्टून, रस्सी, नेवारी)": df[amount_col_name][index + 6], 		
        "पहेले का बकाया": df[amount_col_name][index + 7],		
        "GRAND TOTAL/ कुल": df[amount_col_name][index + 8], 		
        "ADVANCE/ अग्रिम": df[amount_col_name][index + 9], 		
        "PAYABLE/ देय": df[amount_col_name][index + 10], 
        }
    }

    return Dictonary
```

```python
final_list = [ ["Party", "TOTAL NO PCS.", 
"TOTAL AMOUNT",	"(मूर्ति)  दुकान",	"(गणेश लक्ष्मी)  दुकान", "LOADING CHARGE (लेबर खर्चा)",		
"TRANSPORTATION (गाड़ी भाड़ा)", "PACKING CHARGES (कार्टून, रस्सी, नेवारी)", 		
"पहेले का बकाया", "GRAND TOTAL/ कुल", 		
"ADVANCE/ अग्रिम", "PAYABLE/ देय",] ]
```

```python
def build_list(bill, df, index):
    res_list = [
        bill, 
        df[piece_count_col_name][index],		
        df[amount_col_name][index + 1],		
        df[amount_col_name][index + 2],	
        df[amount_col_name][index + 3],		
        df[amount_col_name][index + 4],		
        df[amount_col_name][index + 5],		
        df[amount_col_name][index + 6], 		
        df[amount_col_name][index + 7],		
        df[amount_col_name][index + 8], 		
        df[amount_col_name][index + 9], 		
        df[amount_col_name][index + 10], 
    ]
    return res_list;
```

```python

for bill in tqdm(res, desc = "calculating sum of bills"):
    df = pd.read_excel(
     os.path.join(APP_PATH, "new bills", bill),             # Take both APP_NAME and folder name as Args
     engine='openpyxl')
    curr_list = build_list(bill, df, index)
    final_list.append(curr_list)

print("***********************    Calculation complete!!!    ************************************")
print("\n\n\n")
        
```
